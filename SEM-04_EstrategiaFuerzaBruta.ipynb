**(Caso 01).** Supongamos que tenemos que encontrar el subconjunto de una lista de números que sume 10. Si la lista tiene pocos elementos (como 4 o 5), la fuerza bruta probando todas las combinaciones funciona muy bien

```python
# Libreria itertools. Es ideal para realizar combinaciones, permutaciones, productos cartesianos, etc.
import itertools
```

```python
lista = [2, 4, 6, 3, 7, 5, 2, 9, 1]
print(len(lista))
```

```python
for i in itertools.combinations(lista, 3):
  print(i)
```

```python
def subconjunto_suma(lista, objetivo):
  listado = []
  for i in range(1, len(lista)+1):
      for combinacion in itertools.combinations(lista,i): # (2,4), (2,6), ..., (1,10)
        if sum(combinacion) == objetivo:
          listado.append(combinacion)
  return listado
```

```python
# Ejecutando la funcion
lista = [2, 4, 6, 3, 7, 5, 2, 9, 1, 10]
suma = 37
resultado = subconjunto_suma(lista, suma)
if resultado:
  # print(resultado)
  print(f'Total de combinaciones:  {len(resultado)}')
  for i in resultado:
    print(i)
else:
  print("No se encontro un subconjunto que sume", suma)
#
```

```python
# Considerando el tiempo promedio de ejecución
import time

lista = [2, 4, 6, 3, 7, 5, 2, 9, 1, -6, -5, -3, 8]
suma = 10

t_inicio = time.time()
resultado = subconjunto_suma(lista, suma)
t_fin = time.time()
tiempo_ejecucion = t_fin - t_inicio
print(f'Tiempo de ejecucion: {tiempo_ejecucion} segundos')
if resultado:
  # print(resultado)
  print(f'Total de combinaciones:  {len(resultado)}')
  for i in resultado:
    print(i)
else:
  print("No se encontro un subconjunto que sume", suma)
```

**(Caso 02). **Un chef está planificando un menú de almuerzo y tiene una lista de ingredientes. Quiere elegir una combinación de 3 ingredientes que juntos sumen exactamente 500 calorías. Como aún no tiene una fórmula para decidir cuál es la mejor combinación, decide usar fuerza bruta y probar todas las combinaciones posibles de 3 ingredientes.

```python
Ingrediente = [
    ('Pollo', 250),
    ('Arroz', 200),
    ('Ensalada', 50),
    ('Pescado', 300),
    ('Frijoles', 250),
    ('Pan', 100)
]
```

```python
from itertools import combinations
Ingrediente = [
    ('Pollo', 250),
    ('Arroz', 200),
    ('Ensalada', 50),
    ('Pescado', 300),
    ('Frijoles', 250),
    ('Pan', 100)
]

# Meta de calorías
meta = 500
print("Combinaciones que suman exactamente 500 calorías:\n")

# Generar todas las combinaciones posibles de 3 ingredientes
for combo in combinations(Ingrediente, 3): # [('Pollo', 250),('Arroz', 200),('Ensalada', 50)], [('Pollo', 250),('Arroz', 200),('Pescado', 300)]
    # Sumar las calorías de los 3 ingredientes
    calorias = sum(item[1] for item in combo) # combo: [('Pollo', 250),('Arroz', 200),('Ensalada', 50)], item: ('Pollo', 250)
    # Verificar si la suma es igual a la meta
    if calorias == meta:
        nombres = [item[0] for item in combo]
        print(f"{nombres} → {calorias} cal")
```

```python
import itertools
def menu_optimo(ingredientes, objetivo):
  soluciones = []
  for combo in itertools.combinations(ingredientes, 3): # [('Pollo', 250),('Arroz', 200),('Ensalada', 50)]
    calorias = sum(cal for _,cal in combo)
    if calorias == objetivo:
      soluciones.append(combo)
  return soluciones
#
#
```

```python
Ingrediente = [
    ('Pollo', 250),
    ('Arroz', 200),
    ('Ensalada', 50),
    ('Pescado', 300),
    ('Frijoles', 250),
    ('Pan', 100)
]
resultado = menu_optimo(Ingrediente, 500)
if resultado:
  print(f'Total de combinaciones:  {len(resultado)}')
  for i in resultado:
    print(i)
else:
  print("No se encontro un subconjunto que sume", 300)
```
